```julia
# Funkcja używana do testowania zadanej metody f wykonując n iteracji
function test(f, n)
    e = f(BigFloat(n))
    Δe = BigFloat(ℯ) - e 
    relative_error = abs(Δe/e)
    accuracy = floor(-log10(relative_error))
    println("n=$n:\n $e\nSignif. digits: $accuracy\n")
end
```

***
Poniższa funkcja wylicza wartość $e$ ze wzoru 
$e = \lim\limits_{n \to \infty}{(1 + \frac{1}{n})^n}$

```julia
function calc_e_lim(n)
    (1 + 1/n)^n
end
```

```julia
for k in 1:10
    test(calc_e_lim, 2^k)
end
```

***
Poniższa funkcja oblicza wartość $e$ przy użyciu wzoru Maclaurina $ e = \sum_{k = 0}^{\infty}\frac{1}{k!} $

```julia
function calc_e_series(n)
    fact = 1.0
    sum = fact
    for k in 1:n
        fact *= k
        sum += 1/fact
    end
    return sum
end
```

```julia
for n in 1:100
    test(calc_e_series, n)
end
```

Wykresy:

```julia
using Plots
plotly()
```

```julia
xs = 0:10
range = map(x -> BigFloat(2.0)^x, xs)
approx = map(x -> calc_e_lim(x), range)
p = plot(
    xs, 
    xticks=0:1:10, 
    yticks=2.0:0.05:3.0, 
    ylims=(2.0, 2.8),
    map(x -> BigFloat(MathConstants.e), xs),
    xlabel="\$log_2\\,n\$", 
    ylabel="\$e\\_calc\\_lim(n)\$", 
    seriestype="line",
    label = "Dokładna wartość e",
    legend = :bottomright
)
scatter!(
    xs,
    approx,
    label = "Przybliżona wartość e"
)
```

```julia
xs = 0:10
approx = map(x -> calc_e_series(BigFloat(x)), xs)
plot(
    xs, 
    xticks=0:1:10, 
    yticks=2.0:0.05:3.0, 
    xlims=(0, 10),
    ylims=(2.0, 2.8),
    map(x -> BigFloat(MathConstants.e), xs),
    xlabel="\$n\$", 
    ylabel="\$e\\_calc\\_series(n)\$", 
    seriestype="line",
    label = "Dokładna wartość e",
    legend = :bottomright
)
scatter!(
    xs,
    approx,
    label = "Przybliżona wartość e"
)
```

Porównanie obu metod w skali 1 do 1:

```julia
plot(
    0:20, 
    xticks=0:2:20, 
    yticks=1.0:0.1:3.0, 
    xlims=(0, 20),
    ylims=(1.0, 2.8),
    [
        map(x -> BigFloat(MathConstants.e), 0:20),
        map(x -> calc_e_lim(BigFloat(x)), 0:20),
        map(x -> calc_e_series(BigFloat(x)), 0:20)
    ],
    xlabel="\$n\$", 
    ylabel="\$\\widetilde{e}\$", 
    label = ["e" "calc_e_lim(n)" "calc_e_series(n)"],
    lw = 1,
    legend=:bottomright
)
```

Porównanie obu metod według ilości iteracji, zakładając, że potęgowanie w Julii ma złożoność O(logn)

```julia
plot(
    0:15, 
    xticks=0:2:15, 
    yticks=1.0:0.1:3.0, 
    xlims=(0, 15),
    ylims=(1.0, 2.8),
    [
        map(x -> BigFloat(MathConstants.e), 0:15),
        map(x -> calc_e_lim(2^BigFloat(x)), 0:15),
        map(x -> calc_e_series(BigFloat(x)), 0:15)
    ],
    xlabel="\$liczba\\,\\,iteracji\$", 
    ylabel="\$\\widetilde{e}\$", 
    label = ["e" "calc_e_lim(2^n)" "calc_e_series(n)"],
    lw = 1,
    legend = :bottomright
)
```
